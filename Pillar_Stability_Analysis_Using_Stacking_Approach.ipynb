```python
# library dependecy
import pandas as pd
from google.colab import data_table
data_table.enable_dataframe_formatter()
import seaborn as sns

# machine learning framework
from pycaret.classification import *

# datatable settings
import logging, sys
logging.disable(sys.maxsize)
from google.colab import output
output.enable_custom_widget_manager()
from google.colab import output
output.disable_custom_widget_manager()
```

```python
# read data
df = pd.read_excel('dataset.xlsx')
df
```

```python
# print data skewness
print("skewness of depth =", df['Depth'].skew())
print("skewness of BW =", df['BW'].skew())
print("skewness of PW =", df['PW'].skew())
print("skewness of MH =", df['MH'].skew())
print("skewness of ratio =", df['Ratio'].skew())
```

```python
# plot class distribution
print(df['Stability'].value_counts())
sns.countplot(x = df['Stability'])
```

```python
model_setup = setup(data=df, target='Stability')
```

```python
# compare a few model
compare_models(sort = 'F1')
```

```python
svm = create_model('svm')
dt = create_model('dt')
knn = create_model('knn')

stacker = stack_models([svm, dt, knn])
```

```python
# model evaluation
model_prediction = predict_model(stacker);
```

```python
# testing on new dataset to prediction
new_dataset = df.copy()
new_dataset.drop(columns = ['Stability'], inplace = True)
predict_new_dataset = predict_model(stacker, data = new_dataset);
predict_new_dataset
```

```python
plot_model(stacker, plot = 'confusion_matrix', plot_kwargs={'percent' : True})
```
